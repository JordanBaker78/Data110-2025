```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
plt.style.use("ggplot")
```

```python
x = ['Aggravated Assault','Burglary','Larceny','Motor Vehicle Theft', 'Murder', 'Rape', 'Robbery']
y = [264.1, 250.7, 1347.2, 318.7, 5.7, 38.0, 66.5]
plt.bar(x, y, label='Offense rates per 100,000 people', color='violet')
plt.title('Crime Rate')
plt.xlabel('Crimes')
plt.ylabel('Rates per 100,000 people')
plt.grid(axis='y', linestyle='-', color='black')
plt.legend()
plt.show()
```

```python
x = ['Aggravated Assault','Burglary','Larceny','Motor Vehicle Theft', 'Murder', 'Rape', 'Robbery']
y = [-3.3, -8.1, -4.9, 12, -12, -9.8, -0.8]
plt.barh(x, y, color='indigo')
plt.title('Change in Crime Rates (2022 vs. 2023)')
plt.xlabel('Crime Rate Change')
plt.ylabel('Crime)')
```

```python
data = {'Crime': ['Aggravated Assault','Burglary','Larceny','Motor Vehicle Theft', 'Murder', 'Rape', 'Robbery'],
        'Rate Change': [-3.3, -8.1, -4.9, 12, -12, -9.8, -0.8]}
df = pd.DataFrame(data)
df_sorted = df.sort_values(by='Rate Change', ascending=True)
colors = ['red', 'yellow', 'blue', 'pink', 'indigo', 'violet', 'green']

plt.barh(df_sorted['Crime'], df_sorted['Rate Change'], color=colors)
plt.title('Change in Crime Rates (2022 vs. 2023)')
plt.xlabel('Crime Rate Change')
plt.ylabel('Crime')
plt.grid(axis='y', linestyle='-', color='black')
```

This horizontal bar graph is better than the vertical bar graph because the elements are easier to read and displaying the change in crime rate instead of the total crime rate gives more context. Displaying the change in crime rate instead of the total crime rate allows viewers to see a comparison from one year to the next instead of just numbers from one year.
