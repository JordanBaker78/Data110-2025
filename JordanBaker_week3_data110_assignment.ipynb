```python
x= np.linspace(0, 4*np.pi, 25)
y= x*np.sin(x)
plt.plot(x, y, '-s', color='red',
         markersize=8, linewidth=3,
         markerfacecolor='black',
         markeredgecolor='blue',
         markeredgewidth=3)
plt.xlim(-2,15)
plt.ylim(-15,10)
plt.show()
```

```python
x=np.linspace(-2*np.pi,2*np.pi,100)
y=x*np.cos(x)
plt.plot(x,y,'r')
plt.scatter(x,y, color='blue')
plt.show()
```

```python
x = np.linspace(-10,100,1000)
plt.plot(x, np.tan(x), label='tan(x)', color='red',
         markersize=5, linewidth=3,
         markerfacecolor='black',
         markeredgewidth=3)
plt.plot(x, np.arctan(x), label='arctab(x)', color='blue',
         markersize=5, linewidth=3,
         markerfacecolor='grey',
         markeredgewidth=3)
plt.xlim(0,10)
plt.ylim(-10,10)
plt.title('tan(x) vs arctan(x)')
plt.xlabel("x")
plt.ylabel("tan(x), arctab(x)");
plt.legend()
plt.show();
```
