#Part 2: Data Visualization

##Task 1

```python
#Boxplot - Analyzing Body Mass by Species: Create a boxplot to compare the distribution of penguin body mass across different species:
plt.figure(figsize=(8, 6))
sns.boxplot(x='species', y='body_mass_g', data=df_no_na)
plt.title('Distribution of Penguin Body Mass by Species')
plt.xlabel('Species')
plt.ylabel('Body Mass (g)')
plt.ylim(2000, 7000)
plt.yticks(np.arange(2000, 7000, step=500))
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

###Interpretation

**Which species has the largest median body mass?**

The species with the largest median body mass is the Gentoo species by a significant margin. The median body mass for the Gentoo species is 5000g
compared to around 3750g ofr the Adelie and Chinstrap species.

**Which species has the widest range in body mass?**

The species with the widest range in body mass is the Gentoo species with a range from about 4000g to just under 6500g

**Are there any outliers? If so, which species have them?**

The only species with outliers is the Chinstrap species one minimum outlier at aroung 2750g and one maximum outlier at around 4750g.
