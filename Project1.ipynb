```python
x = ['2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']
y1 = MD_school_scores['Academic Subjects.English.Average GPA']
y2 = MD_school_scores['Academic Subjects.Mathematics.Average GPA']
y3 = MD_school_scores['Academic Subjects.Foreign Languages.Average GPA']
y4 = MD_school_scores['Academic Subjects.Natural Sciences.Average GPA']
y5 = MD_school_scores['Academic Subjects.Social Sciences/History.Average GPA']
y6 = MD_school_scores['Academic Subjects.Arts/Music.Average GPA']

plt.figure(figsize=(12, 8))
plt.style.use('ggplot')
x_indexes = np.arange(len(x))
plt.bar(x_indexes - 0.0625, y1, width=0.125, label='English', color='red')
plt.bar(x_indexes + 0.0625, y2, width=0.125, label='Math', color='blue')
plt.bar(x_indexes - 0.1875, y3, width=0.125, label='Foreign Language', color='orange')
plt.bar(x_indexes + 0.1875, y4, width=0.125, label='Natural Sciences', color='green')
plt.bar(x_indexes - 0.3125, y5, width=0.125, label='Social Sciences/History', color='yellow')
plt.bar(x_indexes + 0.3125, y6, width=0.125, label='Arts/Music', color='purple')
plt.xticks(ticks=x_indexes, labels=x)
plt.legend()
plt.title('Comparison of Subjects Average GPA 2005-2015')
plt.xlabel('Year')
plt.ylabel('Average GPA')
plt.ylim(3, 4)
plt.show()
```

This graph is essentially the previous six graphs in one so we can more easily see the comnparison between each subject. One thing to notice is the while the average GPA's for the Arts/Music classes are far above any other class the GPA's for the Social Sciences/History class are the second highest every year, although the English GPA's do threaten to overtake them in a few years.

Another thing to notice is that from 2005 to 2007, the GPA's for the Natural Science class are slightly higher than the GPA's for the Foreign Langauge class. But in 2008 they are about equal and after that the GPA's for the Foreign Langauge class remain higher than the Natural Science class.

We can also see that the average trend for total GPA's is slighlty positive though there is some variance.
